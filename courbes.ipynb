```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data =  pd.read_csv('donnees_simulation.csv')
data.head(20)
```

```python
# Tracer la courbe pour un véhicule donné (ex: v_0)
veh_id = 'v_0'
df_veh = data[data['vehicle_id'] == veh_id]

plt.plot(df_veh['time'], df_veh['speed'])
plt.title(f"Évolution de la vitesse - {veh_id}")
plt.xlabel("Temps (s)")
plt.ylabel("Vitesse (m/s)")
plt.grid(True)
plt.show()
```

```python
plt.plot(df_veh['x_position'], df_veh['y_position'])
plt.title(f"Trajectoire spatiale - {veh_id}")
plt.xlabel("Position X (m)")
plt.ylabel("Position Y (m)")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
avg_speed_by_time = data.groupby('time')['speed'].mean()

plt.plot(avg_speed_by_time.index, avg_speed_by_time.values)
plt.title("Vitesse moyenne de tous les véhicules dans le temps")
plt.xlabel("Temps (s)")
plt.ylabel("Vitesse moyenne (m/s)")
plt.grid(True)
plt.show()
```

```python
veh_count_by_time = data.groupby('time')['vehicle_id'].nunique()

plt.plot(veh_count_by_time.index, veh_count_by_time.values)
plt.title("Nombre de véhicules actifs à chaque instant")
plt.xlabel("Temps (s)")
plt.ylabel("Nombre de véhicules")
plt.grid(True)
plt.show()
```
